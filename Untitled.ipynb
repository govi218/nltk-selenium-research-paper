```python
def end_str_fn(pos_list_loc):
    end_strs = []
    for l in pos_list_loc:
        cur_pos = ''
        i = 0
        # iterate till end of contrib para
        while i <= len (l): 
            print(i)
            cur_pos = l[i][1]
            if cur_pos == 'VBN':
                # finish current sentence
                while cur_pos != '.':
                    end_strs += cur_pos + ' '
                    cur_pos = l[i+1][1]
                cur_pos = l[i+1][1]
            i +=1
```
